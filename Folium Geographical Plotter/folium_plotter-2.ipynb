```python
import numpy as np
import pandas as pd
import folium as fo
%matplotlib inline
```

```python
folium_map = fo.Map()
```

```python
vol = fo.FeatureGroup(name="My map")
```

```python
volcano_data = pd.read_csv("239.volcano.csv")
```

```python
lat_vol = list(volcano_data["Latitude"])
lon_vol = list(volcano_data["Longitude"])
name_vol = list(volcano_data["Name"])
```

```python
for lat,lon,name in zip(lat_vol, lon_vol,name_vol):
    vol.add_child(fo.Marker(location=[lat, lon], popup=name, icon=fo.Icon(color="green")))
```

```python
folium_map.add_child(vol)
```
